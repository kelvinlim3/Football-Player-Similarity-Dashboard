```python
# preliminaries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_theme(style='dark', font='Century Gothic')
sns.set(rc={'figure.figsize': (12, 8)})
```

```python
# paths for csv files for each position
CB_DATA_PATH = 'Data/data_cb.csv'
FB_DATA_PATH = 'Data/data_fb.csv'
DM_DATA_PATH = 'Data/data_dm.csv'
M_DATA_PATH = 'Data/data_m.csv'
W_DATA_PATH = 'Data/data_w.csv'
CF_DATA_PATH = 'Data/data_cf.csv'
```

```python
# load dataframes for each position
cb_df = pd.read_csv(CB_DATA_PATH)
fb_df = pd.read_csv(FB_DATA_PATH)
dm_df = pd.read_csv(DM_DATA_PATH)
m_df = pd.read_csv(M_DATA_PATH)
w_df = pd.read_csv(W_DATA_PATH)
cf_df = pd.read_csv(CF_DATA_PATH)

df_dict = {'CB': cb_df,
           'FB': fb_df,
           'DM': dm_df,
           'M': m_df,
           'W': w_df,
           'CF': cf_df
          }
```

What is the shape of each dataframe?

```python
for pos, df in df_dict.items():
    print(f'Shape of {pos} position:\t{df.shape}')
```

What do the columns represent and what are they datatypes?

```python
column_dtypes = list(df_dict.values())[0].dtypes
dtypes_same = True
for pos, df in df_dict.items():
    if not df.dtypes.equals(column_dtypes):
        dtypes_same = False
        print('Not all positions have the same columns')

if dtypes_same:
    print(f'Columns for all positions:\n{column_dtypes}')
```

Are there any inaccurate data (null, empty, ?, -)?

```python
total_inaccurate = 0
for pos, df in df_dict.items():
    total_inaccurate += df.isnull().sum().sum() + df[df==''].count().sum() + df[df=='?'].count().sum() + df[df=='-'].count().sum() 

print(f'There is {total_inaccurate} inaccurate data for all positions.')
```

Are there any duplicate rows?

```python
total_duplicates = 0
for pos, df in df_dict.items():
    total_inaccurate += df[df.duplicated()].count().sum()

print(f'There is {total_duplicates} duplicate data for all positions.')
```

```python
# rename 'Player Name' column to Player Details'
new_cb_df = cb_df.rename(columns={'Player Name': 'Player Details'})
new_fb_df = fb_df.rename(columns={'Player Name': 'Player Details'})
new_dm_df = dm_df.rename(columns={'Player Name': 'Player Details'})
new_m_df = m_df.rename(columns={'Player Name': 'Player Details'})
new_w_df = w_df.rename(columns={'Player Name': 'Player Details'})
new_cf_df = cf_df.rename(columns={'Player Name': 'Player Details'})

# add a column `player_name`
cb_player_names = [row['Player Details'].replace(row['Team ID'], '').replace(str(row['Season']), '').strip() for index, row in new_cb_df.iterrows()]
new_cb_df.insert(1, 'Player Name', cb_player_names)
fb_player_names = [row['Player Details'].replace(row['Team ID'], '').replace(str(row['Season']), '').strip() for index, row in new_fb_df.iterrows()]
new_fb_df.insert(1, 'Player Name', fb_player_names)
dm_player_names = [row['Player Details'].replace(row['Team ID'], '').replace(str(row['Season']), '').strip() for index, row in new_dm_df.iterrows()]
new_dm_df.insert(1, 'Player Name', dm_player_names)
m_player_names = [row['Player Details'].replace(row['Team ID'], '').replace(str(row['Season']), '').strip() for index, row in new_m_df.iterrows()]
new_m_df.insert(1, 'Player Name', m_player_names)
w_player_names = [row['Player Details'].replace(row['Team ID'], '').replace(str(row['Season']), '').strip() for index, row in new_w_df.iterrows()]
new_w_df.insert(1, 'Player Name', w_player_names)
cf_player_names = [row['Player Details'].replace(row['Team ID'], '').replace(str(row['Season']), '').strip() for index, row in new_cf_df.iterrows()]
new_cf_df.insert(1, 'Player Name', cf_player_names)

new_df_dict = {'CB': new_cb_df,
               'FB': new_fb_df,
               'DM': new_dm_df,
               'M': new_m_df,
               'W': new_w_df,
               'CF': new_cf_df
              }
```

Do player change primary position?
- No player has changed their primary position

```python
unique_ids = ['Player Name', 'Nationality', 'DOB']
primary_positions_combinations = []
for pos, df in new_df_dict.items():
    uniques = []
    for index, row in df[unique_ids].iterrows():
        if row[unique_ids].tolist() not in uniques:
            uniques.append(row[unique_ids].tolist())

            player_name = row['Player Name']
            nationality = row['Nationality']
            dob = row['DOB']

            player_df = df[(df['Player Name']==player_name) & 
                           (df['Nationality']==nationality) & 
                           (df['DOB']==dob)]
            
            primary_positions = list(set(player_df['Primary Position']))
            
            if primary_positions not in primary_positions_combinations:
                primary_positions_combinations.append(primary_positions)

print(f'All combination of primary positions by individual players:\n{primary_positions_combinations}')
```

What about player who have been transferred or loaned mid-season? Or did not meet the minimum number of appearances? Or played for a team not in the top 5 leagues?

What are the different combination of seasons for each individual player?
- When players are transferred or loaned mid-season, another row is added with the new club and new stats. For instance, Dan Burn was transferred from Brighton to Newcastle United in the middle of the 2021 season, so there are two rows for Dan Burn in 2021, one at Brighton with 13 apps and one at Newcastle United with 16 apps.
- When players do not meet the minimum number of appearances or have played in a team not in the top 5 leagues, they are not included in the data. For instance, Felipe Anderson has 25 appearances for West-Ham United in 2019 and 38 appearances for Lazio in 2021, but in 2020, he only has 2 appearances for West-Ham United and was transferred to FC Porto who play in the Portuguese league mid-season, therefore there is no row for Felipe Anderson in 2020.

```python
# what are the season combinations per player?

unique_ids = ['Player Name', 'Nationality', 'DOB']
seasons_combinations = []
for pos, df in new_df_dict.items():
    uniques = []
    for index, row in df[unique_ids].iterrows():
        if row[unique_ids].tolist() not in uniques:
            uniques.append(row[unique_ids].tolist())

            player_name = row['Player Name']
            nationality = row['Nationality']
            dob = row['DOB']

            player_df = df[(df['Player Name']==player_name) & 
                           (df['Nationality']==nationality) & 
                           (df['DOB']==dob)]
            
            seasons = player_df['Season'].tolist()
            
            if seasons not in seasons_combinations:
                seasons_combinations.append(seasons)

print(f'All combination of seasons by individual players:\n{seasons_combinations}')
```

```python

def cluster_plot(df, k, position):
    # extract player name
    player_name = df['Player Name'].tolist()

    # dataframe of only traits
    reduced_df = df[list(set(RATING_TRAITS+SCORING_TRAITS+CREATING_TRAITS+PASSING_TRAITS+DEFENDING_TRAITS))]

    # PCA for 2D visualisation
    pca = PCA(n_components=2)
    pca_df = pd.DataFrame(pca.fit_transform(reduced_df))
    pca_df.columns = ['pc_1', 'pc_2']

    # perform k-means clustering
    k_means = KMeans(n_clusters=k)
    k_means = k_means.fit(pca_df)
    labels = k_means.predict(pca_df)

    # add cluster and name to dataframe
    pca_df['cluster'] = labels
    pca_df['player_name'] = player_name

    # plot
    ax = sns.lmplot(data=pca_df,
                    x='pc_1', 
                    y='pc_2', 
                    hue='cluster',
                    legend=False, 
                    fit_reg=False, 
                    height=15, 
                    scatter_kws={'s': 250})

    texts = []
    for pc_1, pc_2, player_name in zip(pca_df.pc_1, pca_df.pc_2, pca_df.player_name):
        texts.append(plt.text(pc_1, pc_2, player_name, fontsize=10))

    plt.xlabel('PC 1') 
    plt.ylabel('PC 2')
    plt.title(f'Cluster plot for {position}', size=20)
    plt.show()
```
